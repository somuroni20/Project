ML Algorithms

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
#Feature & Target

x=df1.drop("Deal Valuation",axis=1)
y=df1["Deal Valuation"]
```

```python
x
```

```python
y
```

Linear Regression

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred=lr.predict(x_train)
```

```python
y_pred
```

```python
y_predtest=lr.predict(x_test)
```

```python
y_predtest
```

```python
mse=mean_squared_error(y_train,y_pred)
r2score=r2_score(y_train,y_pred)

print("mse:",mse)
print("r2_score:",r2score)
```

```python
mse1=mean_squared_error(y_test,y_predtest)
r2score1=r2_score(y_test,y_predtest)

print("mse:",mse1)
print("r2_score:",r2score1)
```

Regularization Underfit Algorithm

```python
from sklearn.linear_model import Lasso #L1
```

```python
l1=Lasso(150)
l1.fit(x_train,y_train)
```

```python
l1.score(x_train,y_train)
```

```python
l1.score(x_test,y_test)
```

```python
l1.coef_
```

```python
l1.intercept_
```

```python
for al in range(100,500,50):
    l1=Lasso(al)
    l1.fit(x_train,y_train)
    trainscore=l1.score(x_train,y_train)
    testscore=l1.score(x_test,y_test)
    print("Alpha",al)
    print("train score:",trainscore)
    print("test score:",testscore)
    print("------------------------------------")
```

```python
y_pred1=l1.predict(x_train)
```

```python
y_pred1
```

```python
y_predtest1=l1.predict(x_test)
```

```python
y_predtest1
```

```python
msetrain=mean_squared_error(y_train,y_pred1)
r2scoretrain=r2_score(y_train,y_pred1)
print("mse on training:",msetrain)
print("r2 score in training:",r2scoretrain)
```

```python
msetest=mean_squared_error(y_test,y_predtest1)
r2scoretest=r2_score(y_test,y_predtest1)
print("mse on testing:",msetest)
print("r2 score in testing:",r2scoretest)
```
